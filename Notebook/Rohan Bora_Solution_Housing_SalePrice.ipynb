```python
import pandas as pd
from warnings import filterwarnings
filterwarnings("ignore")
from pandas import set_option
set_option("display.max_rows",100)
set_option("display.max_rows",100)
A = pd.read_csv("C:/Libraries/training_set.csv")
A.head(10)
```

```python
A.shape
```

```python
A.info()
```

```python
A.isna().sum()
```

```python
for i in A.columns:
    if A[i].dtypes != "object":
        x = A[i].mean()
        A[i] = A[i].fillna(x)
    else:
        x = A[i].mode()[0]
        A[i] = A[i].fillna(x)
```

```python
A.isna().sum()
```

```python
A.nunique()
```

```python
A.corr()["SalePrice"].sort_values()
```

```python
X = A.drop(labels = ["Id","SalePrice"],axis = 1)
Y = A[["SalePrice"]]
```

```python
X1 = X.select_dtypes(include="object")
X2 = X.select_dtypes(exclude="object")
```

```python
X1.head(3)
```

```python
X2.head(3)
```

```python
from sklearn.preprocessing import StandardScaler
scalar = StandardScaler()
X2 = pd.DataFrame(scalar.fit_transform(X2),columns=X2.columns)
```

```python
X2.head(3)
```

```python
outliers = []
for i in X2.columns:
    out = X2[(X2[i]<-3)|(X2[i]>3)].index
    outliers.extend(out)
```

```python
outliers
```

```python
from numpy import unique
outliers = unique(outliers)
outliers
```

```python
X = X.drop(index=outliers,axis=0)
Y = Y.drop(index=outliers,axis=0)
```

```python
X.shape
```

```python
Y.shape
```

```python
X = X.reset_index(drop=True)
Y = Y.reset_index(drop=True)
```

```python
X.index
```

```python
Y.index
```

```python
cat = X1.columns
con = X2.columns
```

```python
cat
```

```python
con
```

```python
X_cat = pd.get_dummies(X[cat])
```

```python
X_cat
```

```python
X_con = X[con]
X_con = pd.DataFrame(scalar.fit_transform(X[con]),columns=con)
X_con.head(3)
```

```python
X.new = X_con.join(X_cat)
```

```python
X.new
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest = train_test_split(X.new,Y,test_size=0.2,random_state=31)
```

```python
from statsmodels.api import OLS, add_constant

ols = OLS(ytrain, add_constant(xtrain))
model = ols.fit()
model.summary()
```

```python
model.pvalues.sort_values()
```

```python
col_to_drop = model.pvalues.sort_values().index[-1]
col_to_drop
```

```python
xtrain = xtrain.drop(labels=[col_to_drop],axis=1)
xtest = xtest.drop(labels=[col_to_drop],axis=1)

xtrain.shape
```

```python
ols = OLS(ytrain,add_constant(xtrain))
model = ols.fit()
model.summary()
```

```python
model.pvalues.sort_values()
```

```python
col_to_drop = model.pvalues.sort_values().index[-1]
col_to_drop
```

```python
dropped_columns = []
rsq_adj = []
for i in range (160):
    xtrain = xtrain.drop(labels=col_to_drop,axis=1)
    xtest = xtest.drop(labels=col_to_drop,axis=1)
    ols = OLS(ytrain,add_constant(xtrain))
    model = ols.fit()
    col_to_drop = model.pvalues.sort_values().index[-1]
    dropped_columns.append(col_to_drop)
    rsq_adj.append(model.rsquared_adj)
    
import matplotlib.pyplot as plt
plt.plot(rsq_adj)
```

```python
from sklearn.linear_model import LinearRegression
lm = LinearRegression()
model = lm.fit(xtrain,ytrain)
```

```python
tr_pred = model.predict(xtrain)
ts_pred = model.predict(xtest)
```

```python
from sklearn.metrics import r2_score
tr_r2 = r2_score(ytrain,tr_pred)
ts_r2 = r2_score(ytest,ts_pred)
```

```python
print(f'Training R2 Score : {tr_r2:.4f}')
print(f'Testing R2 Score  : {ts_r2:.4f}')
```

```python
from sklearn.metrics import mean_squared_error
tr_mse = mean_squared_error(ytrain,tr_pred)
ts_mse = mean_squared_error(ytest, ts_pred)

print(f'Training RMSE : {tr_mse**0.5:.4f}')
print(f'Testing RMSE  : {ts_mse**0.5:.4f}')
```

```python
import numpy as np
lambdas = np.arange(0.1,30,0.1)
```

```python
tr = []
ts = []
for i in lambdas:
    rr = Ridge(alpha=i)
    model = rr.fit(xtrain,ytrain)
    tr_pred = model.predict(xtrain)
    ts_pred = model.predict(xtest)
    tr_mse = mean_squared_error(ytrain,tr_pred)
    ts_mse = mean_squared_error(ytest,ts_pred)
    tr.append(tr_mse)
    ts.append(ts_mse)
    if tr_mse>ts_mse:
        print(f'Optimal value of Lambda is {i}')
        break
```
